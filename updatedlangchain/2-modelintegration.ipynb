### Streaming and Batch

### Streaming

```python
from langchain_openai import ChatOpenAI
model=ChatOpenAI(model="gpt-5.4-nano")
for chunk in model.stream("Why do parrots talk?"):
    print(chunk.text, end="", flush=True)
```

### Batch

```python
from langchain_openai import ChatOpenAI
model=ChatOpenAI(model="gpt-5.4-nano")
response = model.batch([
    "why do parrot have colorful father?",
    "How do airplanes fly",
    "What is quantum computing?"
])
for response in response:
    print(response)
```

```python
from langchain_openai import ChatOpenAI
model=ChatOpenAI(model="gpt-5.4-nano")
response = model.batch([
    "why do parrot have colorful father?",
    "How do airplanes fly",
    "What is quantum computing?"
],
config={
    'max-concurrency': 5, #Limits 5 parallel cals
}
)
for response in response:
    print(response)
```
